```python
import pandas as pd
import requests
# import eventlet
from bs4 import BeautifulSoup
from geopy.geocoders import Nominatim
```

```python
from geopy.exc import GeocoderUnavailable
```

```python
url = 'https://www.apartments.com/indianapolis-in/1/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
response = requests.get(url, headers=headers, timeout=30)
```

```python
soup = BeautifulSoup(response.text)
```

```python
print(soup.prettify())
```

```python
name = []
for i in soup.find_all('span',class_='js-placardTitle title'):
    name.append(i.text.strip())
```

```python
print(len(name))
```

```python
print(name)
```

```python
url = 'https://www.apartments.com/indianapolis-in/2/'
response = requests.get(url, headers=headers, timeout=30)
soup = BeautifulSoup(response.text)
for i in soup.find_all('span',class_='js-placardTitle title'):
    name.append(i.text.strip())
```

```python
print(len(name))
```

```python
for i in range(3,28):
    url = 'https://www.apartments.com/indianapolis-in/{}/'.format(i)
    response = requests.get(url, headers=headers, timeout=30)
    soup = BeautifulSoup(response.text)
    for j in soup.find_all('span',class_='js-placardTitle title'):
        name.append(j.text.strip())
    
```

```python
print(len(name))
```

```python
final_name = [] 

for i in name: 
    if i not in final_name: 
        final_name.append(i)
```

```python
print(len(final_name))
```

```python
address = []
for i in range(1,28):
    url = 'https://www.apartments.com/indianapolis-in/{}/'.format(i)
    response = requests.get(url, headers=headers, timeout=30)
    soup = BeautifulSoup(response.text)
    for j in soup.find_all('div',class_='property-address js-url'):
        address.append(j.text.strip())
        

        
```

```python
print(len(address))

```

```python
len(address)
print(address)


for i in range(1,54):
    address.append(None)      
```

```python
print(len(address))
```

```python

pricing = []

for i in range(1,28):
    url = 'https://www.apartments.com/indianapolis-in/{}/'.format(i)
    response = requests.get(url, headers=headers, timeout=30)
    soup = BeautifulSoup(response.text)
    for j in soup.find_all('p',class_='property-pricing') or soup.find_all('div',class_='price-range'):
        pricing.append(j.text.strip())
```

```python
len(pricing)
```

```python
for i in range(1,4):
    pricing.append(-1)

```

```python
long1 = []
lat1 = []
try:
    for i in address:
        location = locator.geocode(i)  # Geocode the current address
        lat1.append(location.latitude)
        long1.append(location.longitude)
except NameError:
    long1.append(None)
    lat1.append(None)  
except AttributeError:
    long1.append(None)
    lat1.append(None)   
except GeocoderUnavailable:
    long1.append(None)
    lat1.append(None) 
        
```
